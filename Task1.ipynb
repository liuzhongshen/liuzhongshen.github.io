### 删除特征

```python
#print (data.columns.tolist())

data = data.drop([ 'trade_no', 'bank_card_no', 'id_name', 'student_feature','source', 'latest_query_time','loans_latest_time'],axis = 1)
print(data.info())
#print(data.shape[0])
```

### 数据类型转换
- 直接在python中对数据进行操作

```python
data['reg_preference_for_trad'].value_counts()
for i in range(0, data.shape[0]):
    if data.reg_preference_for_trad[i] == '一线城市':
        data.reg_preference_for_trad[i] = 1
    elif data.reg_preference_for_trad[i] == '二线城市':
        data.reg_preference_for_trad[i] = 2
    elif data.reg_preference_for_trad[i] == '三线城市':
        data.reg_preference_for_trad[i] = 3
    elif data.reg_preference_for_trad[i] == '境外':
        data.reg_preference_for_trad[i] = 4
    elif data.reg_preference_for_trad[i] == '其他城市':
        data.reg_preference_for_trad[i] = 5
print(data['reg_preference_for_trad'].value_counts())
```

- 将数据导入数据库进行操作，因为数据量很小，所以前面的工作都是直接在python中进行的，实际生产过程中都是在数据库中进行操作

```python
# SQL语句
UPDATE data SET reg_preference_for_trad='2' WHERE reg_preference_for_trad like '1';
UPDATE data SET reg_preference_for_trad='3' WHERE reg_preference_for_trad like '2';
UPDATE data SET reg_preference_for_trad='4' WHERE reg_preference_for_trad like '3';
UPDATE data SET reg_preference_for_trad='5' WHERE reg_preference_for_trad like '4';
UPDATE data SET reg_preference_for_trad='6' WHERE reg_preference_for_trad like '5';
```

注：在实际应用中，**一般情况不使用update语句**，尤其是在金融领域，因为update语句无法追溯，可以使用存储过程。此处为了方便直接使用update语句

### 异常值查看

```python
# 通过箱图为后续删除异常值做准备
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' # 矢量图
plt.figure(figsize=(35, 20))
count = 1
for i in data.columns.values:
    plt.subplot(8,11,count)
    data[i].plot(kind = 'box')
    count += 1
```

### 缺失值处理
- 对于某一特征数据缺失大于30%的行或列删除处理

```python
data = data.dropna(axis=0, thresh = data.shape[1]*0.7) # 行
data = data.dropna(axis=1, thresh = data.shape[0]*0.7) # 列,保留这一列，如果data.shape[0]*0.7个数据非空，其余列滤除
# 使用均值替换缺失值
data_clean = data.fillna(data.mean())
data_clean
```

### 数据切分

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data_clean, test_size=0.3, random_state=2018)
train_data.to_csv('training.csv', index=False, header=True)
test_data.to_csv('testing.csv', index=False, header=True)
```
